```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
###Load the trained model ,Scaler pickle file,one hot encoder
model=load_model('model.h5')

##load the encoder and scaler
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)
with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler_1=pickle.load(file)

```

```python
input_data = {
    "CreditScore": 619,
    "Geography":"France",
    "Gender":"Female",
    "Age": 42,
    "Tenure": 2,
    "Balance": 0.0,
    "NumOfProducts": 1,
    "HasCrCard": 1,
    "IsActiveMember": 1,
    "EstimatedSalary": 101348.9
}
```

```python
geo_encoder=onehot_encoder_geo.transform([[input_data['Geography']]])
```

```python
geoencoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geoencoder_df
```

```python
input_df=pd.DataFrame([input_data])
```

```python
input_df.head()
```

```python
input_df['Gender']=label_encoder_gender.transform(input_df['Gender'])
```

```python
input_df=pd.concat([input_df.drop("Geography",axis=1),geoencoder_df],axis=1)
```

```python
input_df
```

```python
print(type(scaler_1))
```

```python
input_scaled=scaler_1.transform(input_df)
```

```python
input_scaled
```

```python
prediction=model.predict(input_scaled)
```

```python
print(prediction[0][0])
```

```python
if(prediction[0][0]>0.5):
    print("The customer is likely to get churn")
else:
    print("The customer is not likely to churn")
```
